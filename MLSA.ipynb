```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
df=pd.read_csv("Employee_HR.csv")
df.isnull().sum()
arr=['hr','marketing']
df1=df.loc[df['Department'].isin(arr)]
print("The Standard Deviation of salaries of employees in HR and Marketing departments is",df1['Salary_INR'].std())
print("The Variance of salaries of employees in HR and Marketing departments is",df1['Salary_INR'].var())
a=df1['Salary_INR'].quantile([0.25,0.75])
iqr=a[0.75]-a[0.25]
print("The Interquartile range of salaries of employees in HR and Marketing departments is",iqr)
df2=df.loc[df['Department'].isin(['IT'])]
min=df2['time_spent_company'].min()
max=df2['time_spent_company'].max()
exp=df2['time_spent_company'].max()-df2['time_spent_company'].min()
print("The Range of experience in the IT Department is",min,"-",max)
exp
df.boxplot(by='Department',column=['time_spent_company'],grid=False,figsize=(15,10))
df3=df.iloc[:,1:-1]
df3=df3.drop('Department',axis=1)
a=list(df3.std())
b=list(df3.columns)
c={a[i]:b[i] for i in range(len(b))}
a.sort()
print("Highest standard deviation is for "+c[a[-1]])
```
